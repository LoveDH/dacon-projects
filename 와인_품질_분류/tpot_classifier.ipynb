```python
import pandas as pd

from tpot import TPOTClassifier
from sklearn.preprocessing import LabelEncoder

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
submission = pd.read_csv('data/sample_submission.csv')
```

```python
train_data, train_target = train.drop(['id', 'quality'],axis=1), train['quality']
test_data = test.drop(['id'],axis=1)
```

```python
le = LabelEncoder()
train_data['type'] = le.fit_transform(train_data['type'])
test_data['type'] = le.transform(test_data['type'])
```

```python
tpot_params = dict({
    'generations':10,
    'population_size':100,
    'scoring': 'accuracy',
    'subsample':1,
    'verbosity':2,
    'random_state':42,
    'n_jobs':-1,
    'use_dask':True,
    'warm_start':True,
    'memory': 'auto',
    'periodic_checkpoint_folder':'./tpot_checkpoint/'
})
```

```python
tpot = TPOTClassifier(**tpot_params)
for i in range(100):
    tpot.fit(train_data, train_target)
    tpot.export(f'tpot_best_pareto/tpot_{i}.py')
    results = tpot.predict(test_data)
    submission['quality'] = results
    submission.to_csv(f'submission/tpot_g{i}.csv', index=False)
```
